# **1. Very High Human Development**

```python
#definir el detaframe VHD
VHD = pd.read_csv(url, skiprows=7, nrows=66)
#mantener las columnas con datos y eliminar las que tienen espacios vacíos
VHD=VHD.drop(columns=['Unnamed: 3','Unnamed: 5','Unnamed: 7','Unnamed: 9','Unnamed: 11','Unnamed: 13'])
#asegurarnos que los datos en cada columna no posean espacios en blanco al inicio y al final
VHD=VHD.applymap(lambda x: x.strip() if isinstance(x,str) else x)
#renombramos las columnas de interés
VHD=VHD.rename(columns={"Unnamed: 0":"HDIRank","VERY HIGH HUMAN DEVELOPMENT":"Country","Unnamed: 2":"HDI","Unnamed: 4":"lexp","Unnamed: 6":"expyscho","Unnamed: 8":"meanyscho","Unnamed: 10":"GNI","Unnamed: 12":"GNI-HDI","Unnamed: 14":"HDIRank2020"})
#verificar si hay paises duplicados
len(set(VHD['Country']))
#son 66, no hay duplicados
```

```python
#verificamos el tipo de valores que poseen las columnas
VHD.info()
#la columna GNI es del tipo object por poseer elementos str, debería ser numérica
```

```python
#verificaremos cuales columnas tienen missing values
VHD[VHD.isna().any(axis=1)]
#no hay missing value
```

```python
#valores extraños que posee GNI usando el regex '(\W)' que indica todos los elementos menos palabras y números, se convierte en lista y se le resta el simbolo "."
set(VHD['GNI'].str.extractall('(\W)').loc[:,0].to_list())-set('.')
#hay comas
```

```python
#reemplazamos cada coma por "" en GNI
VHD['GNI']=VHD['GNI'].str.replace(",",'')
#convertimes GNI a numérico, los errores se ignoran
VHD['GNI']=VHD['GNI'].apply(lambda col:pd.to_numeric(col,errors='ignore'))
#verificamos que las columnas correspondientes sean numéricas
VHD.info()
#dtype ok
```

```python
#identificamos los countries con simbolos extraños
VHD[VHD['Country'].str.contains(r"\w*[\u00C0-\u01DA']\w*")]

```

```python
#reemplazamos por Turkey
VHD.loc[47, 'Country'] = 'Turkey'
print(VHD.loc[47,'Country'])
```

```python
#Uppercase a todas las cells de Country
VHD['Country']=VHD.Country.str.upper()
```

```python
#estadistica de VHD
VHD.describe()
```

```python
VHD.insert(VHD.shape[1], "Clasificacion",  pd.Series(['very_high_human_development' for x in range(VHD.shape[0])]) )
```

# **5. Other Countries and Territories**

```python
OCT=pd.read_csv(url, skiprows=202, nrows=4)
OCT=OCT.drop(columns=['Unnamed: 3','Unnamed: 5','Unnamed: 7','Unnamed: 9','Unnamed: 11','Unnamed: 13'])
OCT=OCT.applymap(lambda x: x.strip() if isinstance(x,str) else x)
OCT=OCT.rename(columns={"Unnamed: 0":"HDIRank","OTHER COUNTRIES OR TERRITORIES":"Country","Unnamed: 2":"HDI","Unnamed: 4":"lexp","Unnamed: 6":"expyscho","Unnamed: 8":"meanyscho","Unnamed: 10":"GNI","Unnamed: 12":"GNI-HDI","Unnamed: 14":"HDIRank2020"})
len(set(OCT['Country']))
#son 4, no hay duplicados
```

```python
#verificamos el tipo de valores que poseen las columnas
OCT.info()
#solo Country debería ser obj. HDIRank, HDI, expycho, meanyscho, GNI, GNI-HDI, HDIRank2020 deberían tener números
```

```python
#vemos que pasa
OCT
```

```python
#completamos el HDIRank con 1,2,3,4
OCT['HDIRank'] = list(range(1, len(OCT['HDIRank']) + 1))
```

```python
#GNI tienes datos en los que en ves de "," debería ir ""
OCT['GNI']=OCT['GNI'].str.replace(",",'')
```

```python
#Una vez que los datos a los que les corresponde ser numericos lo son, convertimos a numérico los elementos, sino coerce
OCT[['HDI','expyscho','meanyscho','GNI','GNI-HDI','HDIRank2020']] = OCT[['HDI','expyscho','meanyscho','GNI','GNI-HDI','HDIRank2020']].apply(pd.to_numeric, errors='coerce')
#Reemplazamos los NaN por el número 0
OCT[['HDI','expyscho','meanyscho','GNI','GNI-HDI','HDIRank2020']]=OCT[['HDI','expyscho','meanyscho','GNI','GNI-HDI','HDIRank2020']].fillna(0)
```
